# VARMAX
Vector AutoRregression with Moveing Average and eXogenous variable.

Must have at least two endogenous variable, so we will use all 4 bedrooms' data (double differenced to make it stationary).

Zipcode has to be the exogenous variable, since all endog variable will be predicted and thus change.

We need to format the dataframe such that we have time series row index, and columns: 1~4BR as the endog variables, and a zipcode column as the exogenous variable.

```python
# define a function to takes in dataframe and Value_name turn into the dataframe format we wanted
def VARMAX_double_diff(df, val_name):
    tdf = pd.DataFrame({
    'Date': pd.date_range('1996-01-01', periods=293, freq='M')
    })
    # add Zipcode as columns and its values to tdf
    for zipcode in list(df.Zipcode.unique()):
        tdf[zipcode] = list(df[df['Zipcode']==zipcode].Value)
    # set Date column as row index
    tdf.set_index('Date', inplace=True)
    # double difference the data
    tdf = tdf.diff().dropna()
    tdf = tdf.diff().dropna()
    # add Date column to df1dd so we can melt it
    tdf['Date'] = pd.date_range('1996-03-31', periods=291, freq='M')
    # melt and sort
    tdf_sorted = pd.melt(tdf, id_vars=['Date'], var_name='Zipcode', value_name=val_name)
    tdf_sorted = tdf_sorted.sort_values(['Date','Zipcode'])
    # zipcode needs to be integers so it can be used in np.array
    tdf_sorted['Zipcode'] = tdf_sorted['Zipcode'].astype('int64')
    # have the Values in integers for easier viewing
    tdf_sorted[val_name] = tdf_sorted[val_name].astype('float64')
    # set the Date as index
    tdf_sorted.set_index('Date', inplace=True)
    return tdf_sorted
```

```python
# create a function to undo diff
def diff_inv(series_diffed, first_element_original):
    cumsum = series_diffed.cumsum()
    series = cumsum.fillna(0) + first_element_original
    series_original = [first_element_original]
    for i in series:
        series_original.append(i)
    return series_original
```

```python
# define a function to takes in dataframe and Value_name turn into the dataframe format similar to VARMAX_ddpred
def VARMAX_dataformat(df, val_name):
    tdf = pd.DataFrame({
    'Date': pd.date_range('1996-01-01', periods=293, freq='M')
    })
    # add Zipcode as columns and its values to tdf
    for zipcode in list(df.Zipcode.unique()):
        tdf[zipcode] = list(df[df['Zipcode']==zipcode].Value)
    # set Date column as row index
    tdf.set_index('Date', inplace=True)
    # add Date column to df1dd so we can melt it
    tdf['Date'] = pd.date_range('1996-03-31', periods=293, freq='M')
    # melt and sort
    tdf_sorted = pd.melt(tdf, id_vars=['Date'], var_name='Zipcode', value_name=val_name)
    tdf_sorted = tdf_sorted.sort_values(['Date','Zipcode'])
    # zipcode needs to be integers so it can be used in np.array
    tdf_sorted['Zipcode'] = tdf_sorted['Zipcode'].astype('int64')
    # have the Values in float64 in order to model
    tdf_sorted[val_name] = tdf_sorted[val_name].astype('float64')
    # set the Date as index
    tdf_sorted.set_index('Date', inplace=True)
    return tdf_sorted
```

```python
# make the 4 regular orignal dataframes, we will use var_rdf as primary by adding column with values from other dfs
var_rdf = VARMAX_dataformat(df1, 'One_BR')
var_rdf2 = VARMAX_dataformat(df2, 'Two_BR')
var_rdf3 = VARMAX_dataformat(df3, 'Three_BR')
var_rdf4 = VARMAX_dataformat(df4, 'Four_BR')

var_rdf['Two_BR'] = list(var_rdf2['Two_BR'])
var_rdf['Three_BR'] = list(var_rdf3['Three_BR'])
var_rdf['Four_BR'] = list(var_rdf4['Four_BR'])
var_rdf.head()
```

```python
# iterate the zipcodes to reverse the double differenced predictions
# create 4 empty dataframes to store data by number of rooms
VARMAX_dpred1 = pd.DataFrame({'Date': pd.date_range('2018-07-31', periods=(25), freq='M')})
VARMAX_dpred2 = pd.DataFrame({'Date': pd.date_range('2018-07-31', periods=(25), freq='M')})
VARMAX_dpred3 = pd.DataFrame({'Date': pd.date_range('2018-07-31', periods=(25), freq='M')})
VARMAX_dpred4 = pd.DataFrame({'Date': pd.date_range('2018-07-31', periods=(25), freq='M')})
for i in list(var_rdf.Zipcode.unique()):
    df2d = VARMAX_ddpred.copy()
    df2d_sub = df2d[df2d['Zipcode']==i]
    var_rdf_sub = var_rdf[var_rdf['Zipcode']==i]
    var_rdf_subd = var_rdf_sub.diff().dropna()
    first_element_diff = var_rdf_subd.iloc[-25]
    VARMAX_dpred1[i] = diff_inv(df2d_sub['One_BR'], first_element_diff['One_BR'])
    VARMAX_dpred2[i] = diff_inv(df2d_sub['Two_BR'], first_element_diff['Two_BR'])
    VARMAX_dpred3[i] = diff_inv(df2d_sub['Three_BR'], first_element_diff['Three_BR'])
    VARMAX_dpred4[i] = diff_inv(df2d_sub['Four_BR'], first_element_diff['Four_BR'])
VARMAX_dpred1.head()
```

```python
# iterate the zipcodes to reverse the differenced predictions to regular values
# create 4 empty dataframes to store data by number of rooms
VARMAX_pred1 = pd.DataFrame({'Date': pd.date_range('2018-06-30', periods=(26), freq='M')})
VARMAX_pred2 = pd.DataFrame({'Date': pd.date_range('2018-06-30', periods=(26), freq='M')})
VARMAX_pred3 = pd.DataFrame({'Date': pd.date_range('2018-06-30', periods=(26), freq='M')})
VARMAX_pred4 = pd.DataFrame({'Date': pd.date_range('2018-06-30', periods=(26), freq='M')})
for i in list(var_rdf.Zipcode.unique()):
    var1d = VARMAX_dpred1[i]
    var2d = VARMAX_dpred2[i]
    var3d = VARMAX_dpred3[i]
    var4d = VARMAX_dpred4[i]
    var_rdf_sub = var_rdf[var_rdf['Zipcode']==i]
    first_element = var_rdf_sub.iloc[-26]
    VARMAX_pred1[i] = diff_inv(var1d, first_element['One_BR'])
    VARMAX_pred2[i] = diff_inv(var2d, first_element['Two_BR'])
    VARMAX_pred3[i] = diff_inv(var3d, first_element['Three_BR'])
    VARMAX_pred4[i] = diff_inv(var4d, first_element['Four_BR'])
VARMAX_pred1.head()
```

```python
# define a function to takes in prediction dataframe and Value_name
def pred_to_df(df, val_name):
    # melt and sort
    df_m = pd.melt(df, id_vars=['Date'], var_name='Zipcode', value_name=val_name)
    df_sort = df_m.sort_values(['Date','Zipcode'])
    # zipcode needs to be integers so it can be used in np.array
    df_sort['Zipcode'] = df_sort['Zipcode'].astype('int64')
    # have the Values in integers for easier viewing
    df_sort[val_name] = df_sort[val_name].astype('int64')
    return df_sort
```
